```python
car_noise_file = Path(r"/Users/akulaz/Documents/BootCamp/GitHub Repositories/-Project-2-Aeris/car_noise_new.csv")
car_noise = pd.read_csv(car_noise_file)
```

```python
car_noise
car_noise_new = car_noise[['brand', "model", "spec", "year", "dB_at_idle", "dB_at_50kmh", "dB_at_80kmh", "dB_at_100kmh", "dB_at_120kmh", "dB_at_140kmh"]]
```

```python
car_noise_new = car_noise_new.dropna(how="any")
car_noise_new
```

```python
car_noise_new.dtypes
```

```python
car_noise_mph = car_noise_new.copy()
car_noise_mph = car_noise_mph.rename(columns = {"dB_at_50kmh": "db_at_31mph", "dB_at_80kmh": "dB_at_50mph", "dB_at_100kmh": "dB_at_62mph","dB_at_120kmh": "dB_at_75mph", "dB_at_140kmh": "dB_at_87mph"})
```

```python
cars = pd.read_csv("Sport Car.csv")
cars = cars.rename(columns={"Car Model": "model"})
cars
```

```python
carsx2 = pd.merge(cars, car_noise_mph, on="model", how="left")
carsx2 = carsx2.rename(columns = {"model": "Car Model", "dB_at_idle": "dB_at_Idle"})
carsx2 = carsx2[["Car Make", "Car Model", "Year", "Engine Size (L)", "0-60 MPH Time (seconds)", "Price (in USD)", "dB_at_Idle","db_at_31mph", "dB_at_50mph", "dB_at_62mph", "dB_at_75mph", "dB_at_87mph"]]
carsx2
```

```python
carsx2.dtypes
```

```python
carsx2_nonulls = carsx2.dropna(how="any")
carsx2 = carsx2_nonulls.reset_index(drop=True)
```

```python
carsx2 = carsx2_nonulls.sort_values("dB_at_62mph")
carsx2.to_csv("Cars(Noise).csv", index=True)
```

```python
car_models = carsx2.groupby(["Car Make", "Car Model"]).mean()
car_models = car_models[["Engine Size (L)", "0-60 MPH Time (seconds)", "Price (in USD)", "dB_at_Idle","db_at_31mph", "dB_at_50mph", "dB_at_62mph", "dB_at_75mph", "dB_at_87mph"]]
car_models                         
```

```python
car_models.to_csv("Car_Models.csv", index=True)
```
